```python
import pandas as pd
import pandas
import numpy as np
import numpy
from datetime import date
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
###model Parameters

Growth_Factor = 1.18
EBITDA_Margin = 0.45
Taxes = 0.21
Cap_Exp = 0.03
Work_Cap = 0.06
EBITDA_Exit = 32.70
Discount_Factor = 0.94
```

```python
dates = pd.date_range(start='2022',end='2032',freq='Y')
dates
```

```python
df = pd.DataFrame(index=dates)
df
```

```python
#Rev growth:
start_rev = 222
Growth_Factor = 1.18

rev_projection = [start_rev*Growth_Factor]

for i in range(1,len(dates)):

     rev_projection.append( rev_projection[i-1]*Growth_Factor ) 

df['Revenue'] = rev_projection
```

```python
df[['Revenue']]
```

```python
df['EBITDA'] = df['Revenue'] * EBITDA_Margin
df
```

```python
df['Taxes'] = df['EBITDA'] * -Taxes
df
```

```python
df['Cap_Exp'] = df['Revenue'] * -Cap_Exp
df
```

```python
df['Work_Cap'] = df['Revenue'] * -Work_Cap
df
```

```python
df['cash_flows'] = df.drop(columns=['Revenue']).sum(axis=1)
df
```

```python
df['Discount_Factor']=[Discount_Factor ** i for i in range(1,len(dates)+1)]      
df
```

```python
df['present_value_fcf']= Discount_Factor * df['cash_flows']
df
```

```python
print(f""""
 
 Our projected Present Value of Projected Future Cash Flows are: ${df['present_value_fcf'].sum().round(0)}.
We are providing revenue guidance in 2032 to be: ${round(df.iloc[-1]["Revenue"] * Growth_Factor)}.
With an estimated EBITDA around: ${round(EBITDA_Exit)}.

The terminal value at this time given a multiple of ${EBITDA_Exit} and the above EBITDA for 2032, would equal:{round(EBITDA_Exit*EBITDA_Exit)}"

""")
```
